```python
 ## Eine Animation zum Verständnis der Darstellung einer Sinuswelle durch einen Phasor (ähnlich den Projektionen der Zeiger einer Uhr)
    
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
%matplotlib widget

fig=plt.figure(figsize=(5,5))
ax0=fig.add_subplot(2,2,3,polar=True)
ax1=fig.add_subplot(2,2,4)
ax2=fig.add_subplot(2,2,1)

wt=-np.linspace(-2*np.pi,0)

def f_polar(i1,phi11,phi21,amp1,amp2):    
    ax0.clear()
    ax1.clear()
    ax2.clear()        
    
    wt1=2*np.pi*i1/200
    phi1=phi11*np.pi/180
    phi2=phi21*np.pi/180
    
    phasor_1=amp1*np.exp(1j*(wt1+phi1))
        
    phasor_2=phasor_1
    
    ax0.plot([0,wt1+phi1],[0,amp1])
    ax0.plot([0,wt1+phi2],[0,amp2])
    ax0.plot([wt1+phi1,np.angle(phasor_2)],[amp1,abs(phasor_2)])
    
    ax1.plot(-(wt1+wt),amp1*np.sin((wt+wt1)+phi1))
    ax1.plot(-(wt1+wt),amp2*np.sin((wt+wt1)+phi2))    
    
    ax2.plot(amp1*np.cos(wt+wt1+phi1),-(wt1+wt))
    ax2.plot(amp2*np.cos(wt+wt1+phi2),-(wt1+wt))
    
    ax0.set_rlim(0,1.2)
    ax1.set_ylim(-1.2,1.2)
    ax2.set_xlim(-1.2,1.2)
    
    ax1.grid()
    ax2.grid()
        
    return

amp1 = widgets.FloatSlider(min=0,max=1,value=1, description='Amplitude 1:')
phi1 = widgets.FloatSlider(min=-90,max=90,value=0,step=15, description='Phase 1:')
amp2 = widgets.FloatSlider(min=0,max=1,value=.5, description='Amplitude 2:')
phi2 = widgets.FloatSlider(min=-90,max=90,value=30,step=15, description='Phase 2:')


i = widgets.Play(
    min=0,max=200,value=0,
    interval=200,
    description="Press play",
    disabled=False)

widgets.interactive(f_polar,i1=i,phi11=phi1,phi21=phi2,amp1=amp1,amp2=amp2)
```

## Gegeben ist die Schaltung, berechnen Sie die Ströme in jeder Masche ($I_\mathrm{1}$,$I_\mathrm{2}$ und $I_\mathrm{3}$).

<img src="03-GLW-Phasor-1.svg" width="600" float="left"/>

$u_\mathrm{e}(t)=100\mathrm{V}~\sin{(500~t + \frac{\pi}{6})}$<br>
$L_\mathrm{1}=4 \mathrm{mH}$,
$L_\mathrm{2}=40 \mathrm{mH}$,
$L_\mathrm{3}=10 \mathrm{mH}$ und
$C_\mathrm{1}=500  \mathrm{\mu H}$

```python
# Lösungsbeispiel mit Anwendung von Python
import numpy as np
import sympy as sp

# Variabeln
I1,I2,I3 = sp.symbols('I1,I2,I3')

# Phasor
# Spannungsquelle
print('## Spannungsquelle ##')
Ue=100*np.exp(1j*np.pi/6)
print(f'Ue={Ue:.4}V')
print(f'Ue=({abs(Ue):.4}<{np.angle(Ue,deg=True):.4}°) V')

# Impedanzen
print('## Impedanzen ##')
L1=4e-3;L2=40e-3;L3=10e-3;C1=500e-6;R1=20.;R2=10.
w=500
ZL1=1j*w*L1
print(f'ZL1=({ZL1:.4})Ohm')
ZL2=1j*w*L2
print(f'ZL2=({ZL2:.4})Ohm')
ZL3=1j*w*L3
print(f'ZL3=({ZL3:.4})Ohm')
ZC1=1/(1j*w*C1)
print(f'ZC1=({ZC1:.4})Ohm')
ZR1=R1
print(f'ZR1=({ZR1:.4})Ohm')
ZR2=R2
print(f'ZR2=({ZR2:.4})Ohm')

X=sp.solve([-Ue+ZL1*I1+ZR1*(I1-I2), # Mashe 1
           R1*(I2-I1)+ZC1*I2+ZL2*(I2-I3), # Mashe 2
           ZL2*(I3-I2)+ZL3*I3+ZR2*I3],[I1,I2,I3]) # Mashe 3

I1=complex(X[I1])
I2=complex(X[I2])
I3=complex(X[I3])

# Ströme
print('## Ströme ##')
print(f'I1=({abs(I1):.4}<{np.angle(I1,deg=True):.4}°) A')
print(f'I2=({abs(I2):.4}<{np.angle(I2,deg=True):.4}°) A')
print(f'I3=({abs(I3):.4}<{np.angle(I3,deg=True):.4}°) A')
```
