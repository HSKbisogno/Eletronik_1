8. Leiten Sie die normalisierten Parameter der Schaltungen her

##### Zeitraum

Angegebene homogene Lösung:

$i_\mathrm{ah}(t)=K~\mathrm{e}^{a~t}$

Ableitung:

$\frac{\mathrm{d}i_\mathrm{ah}(t)}{\mathrm{d}t}=K~a~\mathrm{e}^{a~t}$

zweite Ableitung

$\frac{\mathrm{d}i_\mathrm{ah}^2(t)}{\mathrm{d}t^2}=K~a^2~\mathrm{e}^{a~t}$

Differentialgleichung

$\frac{\mathrm{d} i_\mathrm{a}(t)^2}{\mathrm{d}t^2}+\frac{R}{L}~\frac{\mathrm{d} i_\mathrm{a}(t)}{\mathrm{d}t}+\frac{1}{L~C} i_\mathrm{a}(t)=0$

homogene Differentialgleichung

$(K~a^2~\mathrm{e}^{a~t})+\frac{R}{L}~(K~a~\mathrm{e}^{a~t})+\frac{1}{L~C}~(K~\mathrm{e}^{a~t})=0$

$(a^2+\frac{R}{L}~a+\frac{1}{L~C})~K~\mathrm{e}^{a~t}=0$

$(a^2+\frac{R}{L}~a+\frac{1}{L~C})=\frac{0}{K~\mathrm{e}^{a~t}}=0$

Die Wurzeln sind

$a=\frac{1}{2}(-\frac{R}{L} \pm \sqrt{(\frac{R}{L})^2-4~\frac{1}{L~C}})$

-------------------------------------------------------------------
Die Wurzeln können normalisiert werden durch

$a=\omega_\mathrm{n}~(-\xi \pm \sqrt{\xi^2-1} )$

$\omega_\mathrm{n}=\sqrt{\frac{1}{L~C}}$ : Eigenfrequenz [rad/s]

$f_\mathrm{n}=\frac{\omega_\mathrm{n}}{2~\pi}$ : Eigenfrequenz [Hz]

$ \xi=\frac{R}{2}~\sqrt{\frac{C}{L}} $ : Dämpfungsfaktor

Die ungedämpfte Amplitude ist

$I_\mathrm{ao}=\sqrt{\frac{C}{L}}~U_\mathrm{e}$ : ungedämpfte Amplitude

----------------------

$ \xi = 0 $ : Konjugiert imaginäre Wurzeln, ungedämpftes System <br>

$ \xi < 1 $ : Konjugiert komplexe Wurzeln, untergedämpftes System <br>

$ \xi = 1 $ : Gleiche reelle Wurzeln, kritisch gedämpftes System <br>

$ \xi > 1 $ : Unterschiedliche reelle Wurzeln, überdämpftes System

-------------------------------------------------------------------

**ξ=0**  

$a_1 = + \omega_\mathrm{n}~\mathrm{j}\\
a_2 = - \omega_\mathrm{n}~\mathrm{j}$

Anfangsbedingungen

$i_\mathrm{a}(0)=0$

$\frac{\mathrm{d} i_\mathrm{a}(0)}{\mathrm{d}t}=\frac{1}{L}~U_\mathrm{e}$

-------------------------------------------------------------------

$i_\mathrm{ah}(0)=K_1~\mathrm{e}^{+ \omega_\mathrm{n}~\mathrm{j}~0}+K_2~\mathrm{e}^{- \omega_\mathrm{n}~\mathrm{j}~0} = 0$

$\frac{\mathrm{d} i_\mathrm{a}(0)}{\mathrm{d}t}= \omega_\mathrm{n}~\mathrm{j}~K_1~\mathrm{e}^{+ \omega_\mathrm{n}~\mathrm{j}~0}- \omega_\mathrm{n}~\mathrm{j}~K_2~\mathrm{e}^{- \omega_\mathrm{n}~\mathrm{j}~0} = \frac{1}{L}~U_\mathrm{e}$

-------------------------------------------------------------------

$K_1+K_2 = 0$

$\omega_\mathrm{n}~\mathrm{j}~K_1- \omega_\mathrm{n}~\mathrm{j}~K_2 = \frac{1}{L}~U_\mathrm{e}$

$K_2=-K_1$

$\omega_\mathrm{n}~\mathrm{j}~K_1+ \omega_\mathrm{n}~\mathrm{j}~K_1 = \frac{1}{L}~U_\mathrm{e}$

$K_1= \frac{1}{2~L~\omega_\mathrm{n}~\mathrm{j}}~U_\mathrm{e} = -\frac{1}{2}~\sqrt{\frac{C}{L}}~\mathrm{j}~U_\mathrm{e}$

$K_2= +\frac{1}{2}\sqrt{\frac{C}{L}}~\mathrm{j}~U_\mathrm{e}$

Euler

| $ \sin(t) =\frac{\mathrm{e}^{\mathrm{j}~t}-\mathrm{e}^{-\mathrm{j}~t}}{2~\mathrm{j}}$ |

$i_\mathrm{ah}(t)=\sqrt{\frac{C}{L}}~U_\mathrm{e}~(\frac{\mathrm{e}^{+\omega_\mathrm{n}~\mathrm{j}~t}-\mathrm{e}^{-\omega_\mathrm{n}~\mathrm{j}~t}}{2~\mathrm{j}}) $

-------------------------------------------------------------------

$i_\mathrm{ah}(t)= I_\mathrm{ao}~\sin(\omega_\mathrm{n}~t)$

-------------------------------------------------------------------

$\frac{\mathrm{d} i_\mathrm{a}(0)}{\mathrm{d}t}= (\sigma + \omega_\mathrm{n}~\mathrm{j})~K_1~\mathrm{e}^{(\sigma + \omega_\mathrm{n}~\mathrm{j})~0}+ (\sigma - \omega_\mathrm{n}~\mathrm{j})~K_2~\mathrm{e}^{(\sigma - \omega_\mathrm{n}~\mathrm{j})~0} = \frac{1}{L}~U_\mathrm{e}$

-------------------------------------------------------------------

$K_2=-K_1$

$K_1= \frac{1}{2~L~\omega_\mathrm{n}~\mathrm{j}}~U_\mathrm{e} = -\frac{1}{2}~\sqrt{\frac{C}{L}}~\mathrm{j}~U_\mathrm{e}$

$K_2= +\frac{1}{2}\sqrt{\frac{C}{L}}~\mathrm{j}~U_\mathrm{e}$

$\frac{\mathrm{d} i_\mathrm{a}(0)}{\mathrm{d}t}= \sigma_1~K_1~\mathrm{e}^{\sigma_1~0}+ \sigma_2~K_2~\mathrm{e}^{\sigma_2~0} = \frac{1}{L}~U_\mathrm{e}$

-------------------------------------------------------------------

$K_1= \frac{1}{L~(\sigma_1-\sigma_2)}~U_\mathrm{e} = \frac{1}{L~(\omega_\mathrm{n}~(-\xi + \sqrt{\xi^2-1} )-\omega_\mathrm{n}~(-\xi - \sqrt{\xi^2-1} ))}~U_\mathrm{e}=\frac{1}{L~(\omega_\mathrm{n}~(-\xi + \sqrt{\xi^2-1} )-\omega_\mathrm{n}~(-\xi - \sqrt{\xi^2-1} ))}~U_\mathrm{e}=\frac{1}{2~L~(\omega_\mathrm{n}~\sqrt{\xi^2-1})}~U_\mathrm{e}$

$K_2= -\frac{1}{2~L~(\omega_\mathrm{n}~\sqrt{\xi^2-1})}~U_\mathrm{e}$

-------------------------------------------------------------------

##### Frequenzraum

$\omega_\mathrm{n}=\sqrt{\frac{1}{L~C}}$ : Eigenfrequenz [rad/s]

$\omega_\mathrm{n}=2~\pi~f_\mathrm{n}$

$ \xi=\frac{R}{2}~\sqrt{\frac{C}{L}} $ : Dämpfungsfaktor

$a=\omega_\mathrm{n}~(-\xi \pm \sqrt{\xi^2-1} )$ : Wurzeln

$BW=f_2-f_1$ : Bandbreite

$f_\mathrm{n}=\sqrt{f_1~f_2}$

$\omega_1=\omega_\mathrm{n}(\sqrt{1+\frac{1}{4~Q_\mathrm{n}^2}}-\frac{1}{2~Q_\mathrm{n}})$ : Grenzfrequenz 1

$\omega_2=\omega_\mathrm{n}(\sqrt{1+\frac{1}{4~Q_\mathrm{n}^2}}+\frac{1}{2~Q_\mathrm{n}})$ : Grenzfrequenz 2

$\omega_1=2~\pi~f_1 \\ \omega_2=2~\pi~f_2$

$B=\frac{\omega_\mathrm{n}}{Q_\mathrm{n}}=\frac{R}{L}$

$BW=\frac{f_\mathrm{n}}{Q_\mathrm{n}}$

$Q_\mathrm{n}=\frac{\omega_\mathrm{n}}{\omega_2-\omega_1}=\frac{f_\mathrm{n}}{f_2-f_1}=\frac{f_\mathrm{n}}{BW}$ : Gütefaktor

$Q_\mathrm{n}=\frac{L~\omega_\mathrm{n}}{R}=\frac{1}{C~\omega_\mathrm{n}R}$

$Q_\mathrm{n}=\sqrt{\frac{L}{C}}\frac{1}{R}=\frac{1}{2~\xi}$

$ \tau = \frac{1}{\xi~\omega_\mathrm{n}} $ : Zeitkostant

$T_{1 \%} = 5~\tau $ : Einschwingzeit (<1%)

$T_{5 \%} = 3~\tau $ : Einschwingzeit (<5%)
