## Aufgaben (d)

8. Leiten Sie die normalisierten Parameter der Schaltungen her

Zeitraum-Lösung der Differenzialgleichung

Angegebene homogene Lösung:

$i_\mathrm{ah}(t)=K~\mathrm{e}^{a~t}$

Ableitung:

$\frac{\mathrm{d}i_\mathrm{ah}(t)}{\mathrm{d}t}=K~a~\mathrm{e}^{a~t}$

Differentialgleichung

$L~\frac{\mathrm{d} i_\mathrm{a}(t)}{\mathrm{d}t}+R~i_\mathrm{a}(t)-U_\mathrm{e}=0$

homogene Differentialgleichung

$L~(K~a~\mathrm{e}^{a~t})+R~(K~\mathrm{e}^{a~t})=0$

$L~(a)+R~(1)=\frac{0}{K~\mathrm{e}^{a~t}}$

$a=-\frac{R}{L}$

Lösung einer Differenzialgleichung

Angegebene Partikuläre Lösung:

$i_\mathrm{a1}(t)=K_1$

Ableitung:

$\frac{\mathrm{d}i_\mathrm{a1}(t)}{\mathrm{d}t}=0$

Differentialgleichung

$L~\frac{\mathrm{d} i_\mathrm{a}(t)}{\mathrm{d}t}+R~i_\mathrm{a}(t)-U_\mathrm{e}=0$

erzwungene Differentialgleichung

$L~(0)+R~(K_1)=U_\mathrm{e}$

$K_1=\frac{U_\mathrm{e}}{R}$

Lösung:

$i_\mathrm{a}(t)=i_\mathrm{ah}(t)+i_\mathrm{a1}(t)$

$i_\mathrm{a}(t)=K~\mathrm{e}^{a~t}+\frac{U_\mathrm{e}}{R}$

$i_\mathrm{a}(0)=0$

$0=K~\mathrm{e}^{a~0}+\frac{U_\mathrm{e}}{R}$

$K=-\frac{U_\mathrm{e}}{R}$

$i_\mathrm{a}(t)=\frac{U_\mathrm{e}}{R}~(1-\mathrm{e}^{-\frac{R}{L}~t})$

-------------------------------------------------------------------

$i_\mathrm{a}(t)=I_\mathrm{o}~(1-\mathrm{e}^{-\frac{1}{\tau}~t})$

-------------------------------------------------------------------

Normalizierte Parameter

$I_\mathrm{o}=\frac{U_\mathrm{e}}{R}$

$\tau=\frac{L}{R}$
