## Aufgaben (d)

8. Leiten Sie die normalisierten Parameter der Schaltungen her

##### Zeitraum

von der [aufgaben 5](../1-d-05/00-ELE1-1-d-05.ipynb) oder von [Zeitraum-Lösung der Differenzialgleichung](00-ELE1-1-d-08_a.ipynb)

$i_\mathrm{a}(t)=\frac{U_\mathrm{e}}{R}~(1-\mathrm{e}^{-\frac{R}{L}~t})$

Dann kann die Lösung normalisiert werden

-------------------------------------------------------------------

$i_\mathrm{a}(t)=I_\mathrm{o}~(1-\mathrm{e}^{-\frac{1}{\tau}~t})$

-------------------------------------------------------------------

Normalizierte Parameter

$I_\mathrm{o}=\frac{U_\mathrm{e}}{R}$

$\tau=\frac{L}{R}$

##### Frequenztraum

Von den [Aufgaben 6](../1-d-06/00-ELE1-1-d-06.ipynb) kann die folgende Übertragungsfunktion normalisiert werden

$G(s)=\frac{1}{R}~\frac{1}{(\frac{L}{R}~s+1)}$

-------------------------------------------------------------------

$G(s)=K~\frac{1}{(\frac{s}{\omega_\mathrm{g}}+1)}$

-------------------------------------------------------------------

Verstärkung im Durchlassbereich

$K=\frac{1}{R}$

oder die Verstärkung im Durchlassbereich in dB

$K_\mathrm{dB} = 20~ \log(K) $

Grenzfrequenz

$\omega_\mathrm{g} = \frac{R}{L}$ in [$\mathrm{rad/s}$]

or

$f_\mathrm{g} = \frac{\omega_\mathrm{g}}{2~\pi}$ in [$\mathrm{Hz}$]
