9. Entwerfen Sie eine Beispielschaltung mittels der normalisierten Parameter im Zeitbereich


```python
# Entwurfbeispiel (Zeitraum)

import numpy as np
import matplotlib.pyplot as plt

########################################################################
# Entwerfen Sie eine Schaltung im Zeitraum mit folgenden Parametern

Ue = 10  # Eingangsspannung
Io = 5   # ungedÃ¤mpfe Schwingungsamplitude
T_1 = 20e-3  # Einschwingzeit
fn=.5*2e3 # Eigenfrequenz [Hz]

########################################################################

# Schaltungsentwurf
# Io = sqrt(C/L) Ue
# C = Io^2 L / Ue^2
# wn = 1/sqrt(L C)
# wn^2 = 1/(L C)
# wn^2 = Ue^2 / (L^2 Io^2)
########################################################################
# L = Ue / (wn Io)
# C = 1/(L wn^2)
########################################################################
# tau = T_1% / 5
# tau = 1 / (wn xi) = 2 L / R
########################################################################
# R = 2 L / tau
########################################################################
wn=2*np.pi*fn
L = Ue / (wn*Io)  # InduktivitÃ¤t der Spule
print(f'L = {L} H')
C = 1/(L*wn**2)  # KapazitÃ¤t des Kondensators"
print(f'C = {C} F')
tau = T_1/5
R = 2*L/tau  # KapazitÃ¤t des Kondensators"
print(f'R = {R} Ohm')

xi = 1/(tau*wn)
print(f'wn = {wn} rad/s')
print(f'xi = {xi} ')
print(f'tau = {tau} s')

t=np.linspace(0,T_1,400)

sigma = - R/(2*L)

ia = np.sqrt(C/L)*Ue*np.exp(sigma*t)*np.sin(wn*t)

ia1 = np.sqrt(C/L)*Ue*np.exp(sigma*t)



plt.figure()
plt.xlabel('$t[\mathrm{s}]$')
plt.ylabel('$i_\mathrm{a}[\mathrm{A}]$')
plt.plot(t,ia)
plt.plot(t,ia1)
plt.plot(t,-ia1)
plt.grid()
plt.show()
```
