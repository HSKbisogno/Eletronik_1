5. Leiten Sie die zeitabhängige Lösung des Stroms $i_\mathrm{a}(t)$ mittels Laplace-Transformation her

Laplace-Transformation Tabelle

| Zeitfunktion - Inverse Laplace-Transformation <br> $f(t)=\mathscr{L}^{-1}\{f(t)\}$| Zeitfunktion - Laplace transformierte Funktion <br> $F(s)=\mathscr{L}\{f(t)\}$ |
|:-------------------------- |:-------------------|
| $f(t)$ | $F(s)$ |
| $\frac{\mathrm{d} f(t)}{\mathrm{d} t}$ | $s~F(s)-f(0)$ |
| $\frac{\mathrm{d} f(t)^2}{\mathrm{d} t^2}$ | $s^2~F(s)-s~f(0)-\frac{\mathrm{d} f(0)}{\mathrm{d} t}$ |
| $1$ | $\frac{1}{s}$ |
| $\mathrm{e}^{a~t}$ | $\frac{1}{s-a}$ |
| $sin(a~t)$ | $\frac{a}{s^2+a^2}$ |
| $cos(a~t)$ | $\frac{s}{s^2+a^2}$ |
| $sin(a~t+b)$ | $\frac{s~sin(b)+a~cos(b)}{s^2+a^2}$ |
| $\mathrm{e}^{a~t}~sin(b~t)$ | $\frac{b}{(s-a)^2+b^2}$ |
| $\mathrm{e}^{a~t}~cos(b~t)$ | $\frac{s-a}{(s-a)^2+b^2}$ |
| $\mathrm{e}^{a~t}~(sin(b~t)+cos(b~t))$ | $\frac{(s-a)+b}{(s-a)^2+b^2}$ |

Differenzialgleichung

$\frac{\mathrm{d} i_\mathrm{a}(t)^2}{\mathrm{d}t^2}+\frac{R}{L}~\frac{\mathrm{d} i_\mathrm{a}(t)}{\mathrm{d}t}+\frac{1}{L~C} i_\mathrm{a}(t)=0$

Mit der Laplace-Tabelle kann man von der Differenzialgleichung zur Laplace-Gleichung überleiten

$s^2~ I_\mathrm{a}(s)-s~i_\mathrm{a}(0)-\frac{\mathrm{d} i_\mathrm{a}(0)}{\mathrm{d}t}+\frac{R}{L}~(s~I_\mathrm{a}(s)-i_\mathrm{a}(0))+\frac{1}{L~C}~I_\mathrm{a}(s)=0$

$I_\mathrm{a}(s)$ muss in der Gleichung isoliert werden

$(s^2+\frac{R}{L}~s+\frac{1}{L~C})I_\mathrm{a}(s)=(s+\frac{R}{L})~i_\mathrm{a}(0)+\frac{\mathrm{d} i_\mathrm{a}(0)}{\mathrm{d}t}$

$I_\mathrm{a}(s)=\frac{(s+\frac{R}{L})~i_\mathrm{a}(0)+\frac{\mathrm{d} i_\mathrm{a}(0)}{\mathrm{d}t}}{(s^2+\frac{R}{L}~s+\frac{1}{L~C})}$

Anfangsbedingungen

$i_\mathrm{a}(0)=i_\mathrm{L}(0)=0$

$u_\mathrm{C}(0)=0$

$L~\frac{\mathrm{d} i_\mathrm{a}(0)}{\mathrm{d}t}+u_\mathrm{C}(0)+R~i_\mathrm{a}(0)-U_\mathrm{e}=0$

$\frac{\mathrm{d} i_\mathrm{a}(0)}{\mathrm{d}t}=\frac{1}{L}~U_\mathrm{e}+\frac{1}{L}~u_\mathrm{C}(0)+\frac{R}{L}~i_\mathrm{a}(0)$

$\frac{\mathrm{d} i_\mathrm{a}(0)}{\mathrm{d}t}=\frac{1}{L}~U_\mathrm{e}$

-------------------------------------------------------------------

$I_\mathrm{a}(s)=\frac{\frac{1}{L}~U_\mathrm{e}}{(s^2+\frac{R}{L}~s+\frac{1}{L~C})}$

-------------------------------------------------------------------

$I_\mathrm{a}(s)=\frac{\frac{1}{L}~U_\mathrm{e}}{(s^2+2~\xi~\omega_\mathrm{n}~s+\omega_\mathrm{n}^2)}$

$\xi=\frac{R}{2}~\sqrt{\frac{C}{L}}$: Dämpfungsfaktor

$\omega_\mathrm{n}=\sqrt{\frac{1}{L~C}}$ : Eigenfrequenz [rad/s]

$r=\omega_\mathrm{n}~(-\xi \pm \sqrt{\xi^2-1} )$ : Wurzeln

-------------------------------------------------------------------

$I_\mathrm{a}(s)=\frac{\frac{1}{L}~U_\mathrm{e}}{(s^2+\frac{R}{L}~s+\frac{1}{L~C})}$

-------------------------------------------------------------------

$I_\mathrm{a}(s)=\frac{U_\mathrm{e}}{L}~\frac{1}{(s-r_1)(s-r_2)}$

$I_\mathrm{a}(s)=\frac{U_\mathrm{e}}{L}~(\frac{A}{(s-r_1)}+\frac{B}{(s-r_2)})$

$A = \lim_{s\to r_1} (s-r_1)~\frac{1}{(s-r_1)(s-r_2)} = \frac{1}{r_1-r_2} $

$B = \lim_{s\to r_2} (s-r_2)~\frac{1}{(s-r_1)(s-r_2)} = \frac{1}{r_2-r_1} $

$I_\mathrm{a}(s)=\frac{U_\mathrm{e}}{L~(r_1-r_2)}~(\frac{1}{(s-r_1)}-\frac{1}{(s-r_2)})$

Mit der Laplace-Tabelle kann man von der Laplace-Transformierten zur Zeitverlaufsgleichung überleiten

$i_\mathrm{a}(t)= \frac{U_\mathrm{e}}{L~(r_1-r_2)}~(\mathrm{e}^{r_1~t}-\mathrm{e}^{r_2~t})$

$(r_1-r_2)=2~\omega_\mathrm{n}~\sqrt{\xi^2-1}$
