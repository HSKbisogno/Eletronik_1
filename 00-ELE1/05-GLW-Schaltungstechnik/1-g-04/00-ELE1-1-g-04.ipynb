4.Leiten Sie die Differenzialgleichungen im Zustandsraum ab

Zustandsraumgleichungen

$ \frac{\mathrm{d} \bf{X}(t)}{\mathrm{d}t}=\bf{A}~\bf{X}(t)+\bf{B}~\bf{U}(t)\\
\bf{Y}(t)=\bf{C}~\bf{X}(t)+\bf{D}~\bf{U}(t)$

Differenzialgleichung

$\frac{\mathrm{d} i_\mathrm{L}(t)}{\mathrm{d}t}=-\frac{R}{L}~i_\mathrm{L}(t)-\frac{1}{L}~u_\mathrm{L}(t)+\frac{U_\mathrm{e}}{L}\\
\frac{\mathrm{d} u_\mathrm{C}(t)}{\mathrm{d}t}=\frac{1}{C}~i_\mathrm{L}(t)$

$i_\mathrm{L}(t)$: Strom der Spule

$u_\mathrm{C}(t)$: Spannung des Kondensators

Zustandsmatrix

$ \bf{X}(t)=\begin{pmatrix} i_\mathrm{L}(t) \\ u_\mathrm{C}(t) \end{pmatrix}$

Eingabematrix

$ \bf{U}(t)=\begin{pmatrix} U_\mathrm{e} \end{pmatrix}$

Ausgabematrix

$ \bf{Y}(t)=\begin{pmatrix} i_\mathrm{a}(t) \end{pmatrix}$

A-Matrix: Systemmatrix

$\bf{A}=\begin{pmatrix} -\frac{R}{L} & -\frac{1}{L} \\ \frac{1}{C} & 0 \end{pmatrix}$

B-Matrix: Eingabematrix

$\bf{B}=\begin{pmatrix} \frac{1}{L} \\ 0 \end{pmatrix}$

C-Matrix: Ausgabematrix

$\bf{C}=\begin{pmatrix} 1 & 0 \end{pmatrix}$

D-Matrix: Durchgangsmatrix

$\bf{D}=\begin{pmatrix} 0 \end{pmatrix}$

Zustandsraumgleichungen

$ \frac{\mathrm{d} }{\mathrm{d}t}\begin{pmatrix} i_\mathrm{L}(t) \\ u_\mathrm{C}(t)\end{pmatrix}=\begin{pmatrix} -\frac{R}{L} & -\frac{1}{L} \\ \frac{1}{C} & 0 \end{pmatrix}~\begin{pmatrix} i_\mathrm{L}(t) \\ u_\mathrm{C}(t)\end{pmatrix}+\begin{pmatrix} \frac{1}{L} \\ 0 \end{pmatrix}~\begin{pmatrix} U_\mathrm{e} \end{pmatrix}\\
\begin{pmatrix} i_\mathrm{a}(t) \end{pmatrix}=\begin{pmatrix} 1 & 0 \end{pmatrix}~\begin{pmatrix} i_\mathrm{L}(t) \\ u_\mathrm{C}(t)\end{pmatrix}+\begin{pmatrix} 0 \end{pmatrix}~\begin{pmatrix} U_\mathrm{e} \end{pmatrix}$
