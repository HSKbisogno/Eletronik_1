## Aufgaben
### Gegeben sind die Schaltungen:

1. Skizzieren Sie den Zeitverlauf
2. Skizzieren Sie den Frequenzverlauf der Übertragungsfunktion $G(s)=\frac{I_\mathrm{a}(s)}{U_\mathrm{e}(s)}$ oder $G(s)=\frac{U_\mathrm{a}(s)}{I_\mathrm{e}(s)}$
3. Leiten Sie die Differenzialgleichungen ab
4. Leiten Sie die Differenzialgleichungen im Zustandsraum ab (Die Ausgangvariable ist $i_\mathrm{a}(t)$)
5. Leiten Sie die zeitabhängige Lösung des Stroms $i_\mathrm{a}(t)$ mittels Laplace-Transformation her
6. Leiten Sie die Übertragungsfunktion $G(s)=\frac{I_\mathrm{a}(s)}{U_\mathrm{e}(s)}$ oder $G(s)=\frac{U_\mathrm{a}(s)}{I_\mathrm{e}(s)}$ ab 
7. Leiten Sie die Übertragungsfunktion $G(s)=\frac{I_\mathrm{a}(\mathrm{j} \omega)}{U_\mathrm{e}(\mathrm{j} \omega)}$ oder $G(s)=\frac{U_\mathrm{a}(s)}{I_\mathrm{e}(s)}$ mittels Phasor-Gleichungen her
8. Leiten Sie die normalisierten Parameter der Schaltungen her
9. Entwerfen Sie eine Beispielschaltung mittels der normalisierten Parameter im Zeitbereich
10. Entwerfen Sie eine Beispielschaltung mittels der normalisierten Parameter im Frequenzbereich
