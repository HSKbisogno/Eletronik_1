4.Leiten Sie die Differenzialgleichungen im Zustandsraum ab

Zustandsraumgleichungen

$ \frac{\mathrm{d} \bf{X}(t)}{\mathrm{d}t}=\bf{A}~\bf{X}(t)+\bf{B}~\bf{U}(t)\\
\bf{Y}(t)=\bf{C}~\bf{X}(t)+\bf{D}~\bf{U}(t)$

Differenzialgleichung

$C \frac{\mathrm{d}~u_\mathrm{C}(t)}{\mathrm{d}t}=i_\mathrm{a}(t)$<br>

$i_\mathrm{a}(t)=C~U_\mathrm{e}~\delta(t)$<br>
$\delta(t)$: Dirac-Delta-Funktion

oder

$\frac{\mathrm{d}~u_\mathrm{C}}{\mathrm{d}t}=U_\mathrm{e}~\delta(t)$

$u_\mathrm{C}(t)$: Spannung des Kondensators

Zustandsmatrix

$ \bf{X}(t)=\begin{pmatrix} u_\mathrm{C}(t) \end{pmatrix}$

Eingabematrix

$ \bf{U}(t)=\begin{pmatrix} U_\mathrm{e}~\delta(t) \end{pmatrix}$

Ausgabematrix

$ \bf{Y}(t)=\begin{pmatrix} i_\mathrm{a}(t) \end{pmatrix}$

A-Matrix: Systemmatrix

$\bf{A}=\begin{pmatrix} 0 \end{pmatrix}$

B-Matrix: Eingabematrix

$\bf{B}=\begin{pmatrix} 1 \end{pmatrix}$

C-Matrix: Ausgabematrix

$\bf{C}=\begin{pmatrix} 0 \end{pmatrix}$

D-Matrix: Durchgangsmatrix

$\bf{D}=\begin{pmatrix} C \end{pmatrix}$
